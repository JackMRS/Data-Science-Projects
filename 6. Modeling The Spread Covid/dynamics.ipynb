```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def seuvr(x,t):

    # parameters
    beta = 0.000003
    ep = 0.05
    p_v = 0.5
    p_u = 0.5
    d_v = 0.05
    d_u = 0.08
    r_v = 0.10
    r_u = 0.05
    xi = 0.2
    
    #initial conditions
    s = x[0]
    e = x[1]
    i = x[2]
    v = x[3]
    u = x[4]
    r = x[5]
    d = x[6]
    
    dsdt = -beta*s*(i-u)+xi*r
    dedt = beta*s*(i-u)-ep*e
    didt = ep*e-(d_u+r_u)*u-(d_v+r_v)*v
    dvdt = ep*e*p_v-(d_v+r_v)*v
    dudt = ep*e*p_u-(d_u+r_u)*u
    drdt = r_v*v+r_u*u-xi*r
    dddt = d_v*v+d_u*u
    
    return [dsdt,dedt,didt,dvdt,dudt,drdt,dddt]
```

```python
#initial conditions
x0 = [1000000, 0, 5, 4, 1, 0, 0]
# run time
t = np.linspace(0, 300, 200)
#calculate
x = odeint(seuvr,x0,t)

s = x[:,0]
e = x[:,1]
i = x[:,2]
v = x[:,3]
u = x[:,4]
r = x[:,5]
d = x[:,6]
```

```python
plt.figure(figsize=(8,5))
plt.plot(t,s,label="S")
plt.plot(t,e,label="E")
plt.plot(t,i,label="I")
plt.plot(t,v,label="V")
plt.plot(t,u,label="U")
plt.plot(t,r,label="R")
plt.plot(t,d,label="D")
plt.legend()
plt.xlabel("Time",fontsize=20)
plt.ylabel("Compartment Population",fontsize=20)
plt.legend(fontsize=15)
plt.savefig('Population.png')
```
