```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def seuvr(x,t):
    beta = 0.000003
    ep = 0.07
    p_v = 0.7
    p_u = 0.3
    d_v = 0.05
    d_u = 0.08
    r_v = 0.15
    r_u = 0.05
    xi = 0.2
    
    s = x[0]
    e = x[1]
    i = x[2]
    v = x[3]
    u = x[4]
    r = x[5]
    d = x[6]
    
    dsdt = -beta*s*(i-u)+xi*r
    dedt = beta*s*(i-u)-ep*e
    didt = ep*e-(d_u+r_u)*u-(d_v+r_v)*v
    dvdt = ep*e*p_v-(d_v+r_v)*v
    dudt = ep*e*p_u-(d_u+r_u)*u
    drdt = r_v*v+r_u*u-xi*r
    dddt = d_v*v+d_u*u
    
    return [dsdt,dedt,didt,dvdt,dudt,drdt,dddt]
```

```python
x0 = [1e6, 0, 5, 4, 1, 0, 0]
t = np.linspace(0, 200, 200)
x = odeint(seuvr,x0,t)

s = x[:,0]
e = x[:,1]
i = x[:,2]
v = x[:,3]
u = x[:,4]
r = x[:,5]
d = x[:,6]
```

```python
plt.plot(t,s,label="s")
plt.legend()
```
