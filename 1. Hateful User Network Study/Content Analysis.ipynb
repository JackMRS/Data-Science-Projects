```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

data_s=[]

with open("hashtag list.csv","r",encoding='utf-8') as file:
    for line in file:
        data = line.strip().split(",")
        data_s.append(data)

tag =data_s[0]
users=[]
for data in data_s[1:]:
    user=dict(zip(tag,data))
    users.append(user)
```

```python
h_users=[]
n_users=[]
o_users=[]

for user in users:
    if user['hate']=='hateful':
            h_users.append(user)
    if user['hate']=='normal':
            n_users.append(user)
    if user['hate']=='other':
            o_users.append(user)

print(len(h_users))
print(len(n_users))
print(len(o_users))
```

```python
hate=''
normal=''
others=''
for user in h_users:
    hate=hate+user['hashtags']+' '
for user in n_users:
    normal=normal+user['hashtags']+' '
for user in o_users:
    others=others+user['hashtags']+' '
```

```python
h_frequency = {}
for word in hate.split():
    if word not in h_frequency:
        h_frequency[word] = 1
    else:
        h_frequency[word] += 1
h_hashtag_sort=sorted(h_frequency.items(),key=lambda x:x[1],reverse=True) #high to low
print(len(hate.split()))
for hashtag in h_hashtag_sort[0:20]:
    print(hashtag)
    print(str(100*hashtag[1]/25310)+'%')
```

```python
n_frequency = {}
for word in normal.split():
    if word not in n_frequency:
        n_frequency[word] = 1
    else:
        n_frequency[word] += 1
n_hashtag_sort=sorted(n_frequency.items(),key=lambda x:x[1],reverse=True) #high to low
print(len(normal.split()))
for hashtag in n_hashtag_sort[0:20]:
    print(hashtag)
    print(str(100*hashtag[1]/363115)+'%')
```

```python
o_frequency = {}
for word in others.split():
    if word not in o_frequency:
        o_frequency[word] = 1
    else:
        o_frequency[word] += 1
o_hashtag_sort=sorted(o_frequency.items(),key=lambda x:x[1],reverse=True) #high to low
print(len(others.split()))
for hashtag in o_hashtag_sort[0:20]:
    print(hashtag)
    print(str(100*hashtag[1]/8463456)+'%')
```
