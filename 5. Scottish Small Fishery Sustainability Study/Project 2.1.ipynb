```python
import numpy as np
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def det(i):
    arr=[]
    arr.append(i)
    for i in range(20):
        x = arr[-1]*(1 + 0.4)*0.68
        arr.append(math.floor(x))
    return arr

def dem(i):
    arr=[]
    arr.append(i)
    for i in range(20):
        x = x = arr[-1]*(1 + np.random.normal(0.4, 0.1))*np.random.normal(0.68, 0.07)
        arr.append(math.floor(x))
    return arr

def env(i):
    arr=[]
    arr.append(i)
    for i in range(20):
        if (np.random.uniform(0,1)<=0.04):
            x = arr[-1]*(1 + 0.4*(1-0.3))*0.68*(1-0.3)
            arr.append(math.floor(x))
        else:
            x = arr[-1]*(1 + 0.4)*0.68
            arr.append(math.floor(x))
    return arr

def mix(i):
    arr=[]
    arr.append(i)
    for i in range(20):
        if (np.random.uniform(0,1)<0.04):
            x = arr[-1]*(1 + np.random.normal(0.4, 0.1)*(1-0.3))*np.random.normal(0.68, 0.07)*(1-0.3)
            arr.append(math.floor(x))
        else:
            x = arr[-1]*(1 + np.random.normal(0.4, 0.1))*np.random.normal(0.68, 0.07)
            arr.append(math.floor(x))
    return arr
```

```python
y1=det(100)
df1 = pd.DataFrame (y1,columns=['number'])
df1.to_csv(r'111.csv')
plt.plot(y1)
plt.xlabel("Time (s)")
plt.ylabel("Num")
```

```python
y2=[]
for i in range(50):
    y2.append(dem(100))
df2=pd.DataFrame(np.row_stack(y2))
df2.to_csv(r'222.csv')
```

```python
for i in range(50):
    plt.plot(y2[i])

plt.xlabel("Year")
plt.ylabel("Number")
```

```python
y3=[]
for i in range(50):
    y3.append(env(100))
df3=pd.DataFrame(np.row_stack(y3))
df3.to_csv(r'333.csv')

for i in range(50):
    plt.plot(y3[i])

plt.xlabel("Year")
plt.ylabel("Number")
```
