```python
import matplotlib.pyplot as plt
```

```python
t=[]
ca=[]
us=[]

t_0=1960
ca_0 = 15206
us_0 = 163040
t.append(t_0)
ca.append(ca_0)
us.append(us_0)

delta_t=5
for i in range(1,13):
    t.append(t[i-1]+delta_t)
for i in range(1,13):
    caa = ca[i-1]*1.170773
    ca.append(caa)
    uss = us[i-1]*1.0720533
    us.append(uss)
    
```

```python
plt.plot(t,ca,label='California (predict)' )
plt.plot(t,us,label='Rest of U.S. (predict)')
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
```

```python
ca_act=[15870 ,18580 ,19970 ,21540 ,24290 ,29950 ,31490 ,33630 ,35770 ,37270 ,38920 ,39510]
us_act=[183700,194300,205100,216000,226500,237900,250100,266600,282200,295500,309300,328200]

us_rest=[]
for i in range(0,12):
    us_rest.append(us_act[i]-ca_act[i])
```

```python
plt.plot(t,ca,label='California (predict)' )
plt.plot(t,us,label='Rest of U.S. (predict)')
t.pop()
plt.plot(t,ca_act,label='California (actual)' )
plt.plot(t,us_rest,label='Rest of U.S. (actual)')
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
```
