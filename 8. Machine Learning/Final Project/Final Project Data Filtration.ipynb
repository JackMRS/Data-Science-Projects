```python
import pandas as pd
```

```python
users_df = pd.read_csv('users_neighborhood_anon.csv') 
users_df.head()
```

```python
target_users_df = pd.read_csv('HN Edges.csv') 
target_users_df.head()
```

```python
indexes_to_drop = []
for index, row in users_df.iterrows():
    if (row['user_id'] not in target_users_df['source'].to_list()) and (row['user_id'] not in target_users_df['target'].to_list()):
        indexes_to_drop.append(users_df.index[index])
    
users_df.drop(indexes_to_drop, inplace=True)
users_df
```

```python
users_df = users_df[pd.notnull(users_df["negotiate_empath"])]
users_df
```

```python
users_df.to_csv('HN_users_data.csv', index = False) 
```
